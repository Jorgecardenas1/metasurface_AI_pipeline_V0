```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv2d') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

## loading and comparing data

```python
import torch.optim as optimizer

fwd_test = Stack.Predictor_CNN(cond_input_size=parser.condition_len, 
                               ngpu=0, image_size=parser.image_size ,
                               output_size=8, channels=3,
                               features_num=3000,
                               dropout=0.2, 
                               Y_prediction_size=601) #size of the output vector in this case frenquency points

fwd_test.apply(weights_init)

"""using weigth decay regularization"""
opt = optimizer.Adam(fwd_test.parameters(), lr=parser.learning_rate, betas=(0.5, 0.999),weight_decay=1e-4)
criterion = nn.CrossEntropyLoss()
fwd_test.train()
```

```python
import glob
from tqdm.notebook import tqdm


#File reading conf
a = []
idx=0
iters=0

loss_values = []

for epoch in range(parser.epochs):
    x=0
    running_loss = 0.0
    i=0
    print('Epoch {}/{}'.format(epoch, parser.epochs - 1))
    print('-' * 10)
    
    
    dataloader = utils.get_data_with_labels(512, 512,0.9, boxImagesPath,parser.batch_size,drop_last=True)


    for data in tqdm(dataloader):
        
        inputs, classes, names = data
        
        opt.zero_grad()
        #Loading data
        a = []
        idx=0
        
        """lookup for data corresponding to every image in training batch"""
        for name in names:
            series=name.split('_')[-1].split('.')[0]
            batch=name.split('_')[4]
            for name in glob.glob(DataPath+batch+'\\files\\'+'/'+parser.metricType+'*'+series+'.csv'): 
                
                #loading the absorption data
                train = pd.read_csv(name)
                values=np.array(train.values.T)
                a.append(values[1])
        a=np.array(a)     
        
       
        conditioning=torch.ones(1,inputs.shape[0],parser.condition_len)
        
        y_predicted=fwd_test(input_=inputs, conditioning=conditioning, b_size=inputs.shape[0])
        y_predicted=torch.nn.functional.normalize(y_predicted, p=2.0, dim = 1)
       
        y_truth = torch.tensor(a)

        #error
        
        errD_real = criterion(y_predicted.float(), y_truth.float())
        errD_real.backward()
        loss=errD_real.item()
        opt.step()
        scale = torch.tensor([10.0])

        running_loss +=loss*inputs.size(0)


        x += 1
        i = i+1
        

        if i % 10 == 5:    # print every 2000 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {loss / 10:.3f} running loss:  {running_loss / 10:.3f}')

        iters += 1
    
    loss_values.append(running_loss)


```

```python
# Model loading
net = Stack.Predictor_CNN(cond_input_size=parser.condition_len, 
                               ngpu=0, image_size=parser.image_size ,
                               output_size=8, channels=3,
                               features_num=3000,
                               dropout=0.2, 
                               Y_prediction_size=601) #size of the output vector in this case frenquency points
net.eval()
net.load_state_dict(torch.load(PATH))
```

```python

fig, axs = plt.subplots(16,2,figsize=(10,30))

counter=0
    
for item in dataloader_test:
    inputs,target,names = item
    train=[]

    
    for name in names:
        series=name.split('_')[-1].split('.')[0]
        batch=name.split('_')[4]
        for name in glob.glob(DataPath+batch+'\\files\\'+'/'+parser.metricType+'*'+series+'.csv'): 
            #loading the absorption data
            train.append(pd.read_csv(name))

    conditioning=torch.ones(1,inputs.shape[0],parser.condition_len)

    #batch size
    output_test=net(input_=inputs, conditioning=conditioning, b_size=inputs.shape[0])

    output_to_plot=output_test.cpu().detach().numpy()
    images_to_plot = inputs.detach().cpu().permute(0, 2, 3,1).numpy()
    
    
    for i in range(len(train)):
        
        x=np.array(train[i].values.T)[0]
        y_true=np.array(train[i].values.T)[1]
        
        print(names[i])
        
        toPlot=output_to_plot[i]
        
        indices = toPlot < 0.01
        toPlot[indices]=0

        images = (images_to_plot[i] * 255).round().astype("uint8")
        pil_images = images

        axs[counter][0].imshow(pil_images)
        axs[counter][0].set_axis_off()
        
        axs[counter][1].plot(x,y_true,label="Simulated")
        axs[counter][1].plot(x,toPlot,label="Predicted")
        axs[counter][1].legend(loc="upper right")
        axs[counter, 1].set(xlabel='Freq', ylabel='Abs')
        counter=counter+1

  
```
