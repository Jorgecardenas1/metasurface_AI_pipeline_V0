```python
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Linear') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
import glob
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import torch.optim as optimizer


#File reading conf
a = []
idx=0
metricType='AbsorbanceTM'

#Training conf
beta1 = 0.5

optim = optimizer.Adam(dnn.parameters(), lr=parser.learning_rate, betas=(beta1, 0.999),weight_decay=1e-4)
#optim = optimizer.SGD(dnn.parameters(), lr=parser.learning_rate, momentum=0.9)

criterion = nn.BCELoss()

acc=0
acc_test=0
loss=0
test_loss=0
loss_values = []
test_loss_values = []
train_acc_hist = []
test_acc = []

x=0
running_loss = 0.0
loss_values = []
iters=0
i=0
for epoch in range(parser.epochs):
    i=0
    for data in tqdm(dataloader):
        inputs, classes, names = data    
        
        optim.zero_grad()

        #Loading data
        a = []
        idx=0

        flattened= torch.flatten(inputs,1)
        conditioning=torch.ones(1,inputs.shape[0],10)

        input_vector=torch.cat((flattened,conditioning[0]),1)

        a = []
        idx=0


        for name in names:
            series=name.split('_')[-1].split('.')[0]
            batch=name.split('_')[4]
            for name in glob.glob(DataPath+batch+'\\files\\'+'/'+metricType+'*'+series+'.csv'): 
                #loading the absorption data
                train = pd.read_csv(name)
                values=np.array(train.values.T)
                a.append(values[1])

        a=np.array(a)    

        dnn.train()

        output=dnn(input_vector)



        y_truth = torch.tensor(a)
        y_truth = torch.nn.functional.normalize(y_truth, p=2.0, dim = 1)

        #error

        error = criterion(output.float(), y_truth.float())
        error.backward()
        loss=error.item()
        optim.step()

        scale = torch.tensor([10.0])

        running_loss =scale* loss
        loss_values.append(running_loss)

        x += 1

        i=i+1

        #if i % 10 == 9:    # print every 2000 mini-batches
        #print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 10:.3f}')
        #running_loss = 0.0

        iters += 1
    
    
```
